```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
df = pd.DataFrame(iris['data'], columns=['SL','SW','PL','PW'])
df['target'] = iris['target']
df.head()
```

```python
xtr, xts, ytr, yts = train_test_split(
    df[['SL','SW','PL','PW']],
    df['target'], test_size=.1
)
xtr.head()
```

```python
#gabungin xtr dan ytr 

dfTr = pd.concat([xtr, ytr], axis=1)
dfTs = pd.concat([xts, yts], axis=1)
# dfTs
```

```python
#function untuk hitung score

def hitungScore(model, xtr, xts, ytr, yts):
    model.fit(xtr,ytr)
    return model.score(xts,yts)
```

```python
#splitting dataframe training dengan KFold
score=[]
kf = KFold(n_splits=5)
for i, j in kf.split(np.array(dfTr)):
#     print(i,j) #i = index untuk trainingnya, j = index untuk validation
#     print(dfTr.iloc[i])
    xtrcv = dfTr.iloc[i][['SL','SW','PL','PW']]
    ytrcv = dfTr.iloc[i]['target']
    xvalcv = dfTr.iloc[j][['SL','SW','PL','PW']]
    yvalcv = dfTr.iloc[j]['target']
    score.append(hitungScore(LogisticRegression(), xtrcv, xvalcv, ytrcv, yvalcv))
score
np.mean(score)
#dfTr = 135 => 5 fold => 27
#training = 27 + 27 + 27 + 27 = | validation = 27
```

```python
#Cara 2 untuk Cross Validation

print(cross_val_score(
    LogisticRegression(),dfTr[['SL','SW','PL','PW']],
    dfTr['target'], cv=5
))

print(np.mean(cross_val_score(
    LogisticRegression(),dfTr[['SL','SW','PL','PW']],
    dfTr['target'], cv=5, scoring='accuracy'
)))
```

```python
#KNN

model1 = KNeighborsClassifier(n_neighbors=3)
model2 = KNeighborsClassifier(n_neighbors=4)
```

```python
score1 = cross_val_score(model1, dfTr[['SL','SW','PL','PW']],
    dfTr['target'], cv=5, scoring='balanced_accuracy')
print(score1)
print(score1.mean())
```

```python
score2 = cross_val_score(model2, dfTr[['SL','SW','PL','PW']],
    dfTr['target'], cv=5, scoring='balanced_accuracy')
print(score2)
print(score2.mean())
```

# Hyper-parameter tuning

```python
df = pd.DataFrame(iris['data'], columns=['SL','SW','PL','PW'])
df['target'] = iris['target']
df.head()
```

```python
xtr, xts, ytr, yts = train_test_split(
    df[['SL','SW','PL','PW']], df['target']
)
xtr.head()
```

```python
modelAsli = LogisticRegression()
modelAsli.fit(xtr,ytr)
```

```python
modelAsli.score(xts,yts)
```

```python
print(cross_val_score(modelAsli, xtr, ytr))
print(np.mean(cross_val_score(modelAsli, xtr, ytr)))
```

```python
# param yang akan di tuned + nilai yang mungkin
penalty = ['l1','l2','elasticnet','none']
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
max_iter = [1, 10, 100, 1000, 10000]


param = {
    'penalty':penalty, 'solver':solver, 'max_iter':max_iter
}
param
```

```python
# model.get_params()
```

### Hyperparameter Tuning:
- Randomized Search Cross Validation 
- Grid Search Cross Validation

### Randomized Search CV

```python
model = LogisticRegression()
modelrs = RandomizedSearchCV(
    estimator = model, param_distributions = param, cv = 5 
)
```

```python
modelrs.fit(xtr,ytr)
modelrs.best_params_
```

```python
modelAsli.score(xts,yts)
```

```python
modelBaru = LogisticRegression(
    solver='lbfgs', penalty='none', max_iter=10
)
modelBaru.fit(xtr,ytr)
modelBaru.score(xts, yts)
```

### Grid Search CV 

```python
model = LogisticRegression()
modelgs = GridSearchCV(
    model, param, cv=5 
)
```

```python
modelgs.fit(xtr, ytr)
modelgs.best_params_
```

```python
modelAsli.score(xts, yts)
```

```python
modelBaru = LogisticRegression(
    max_iter= 10, penalty= 'l2', solver= 'sag'
)
modelBaru.fit(xtr, ytr)
modelBaru.score(xts, yts)
```
